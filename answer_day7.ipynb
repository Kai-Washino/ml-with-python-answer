### 問題1
くにちゃんは実験で色々なデータを収集しました．  
実験では，被験者5人の700歩分のデータを収集しました．  
700歩は100歩 × 7路面です．  
データは100歩分のデータが一つのファイルで，合計35ファイルあります．  

- くにちゃんはどのような前処理をしたでしょうか？
- くにちゃんはどのような学習をしたでしょうか？
- 被験者内5分割検証，被験者間5分割交差検証をしました．どうやってしたでしょう？

前処理
- まず1歩ずつのデータに分割する．
- その1歩の発電パルスに対して，4つの特徴量を取得する．
- 平均値，分散値，最大値，最小値

学習
- 4つの特徴量をRandomForestにいれる．
- 正解ラベルは路面

交差検証
- 被験者内5分割交差検証は，1路面のデータ80歩分を学習データ，1路面のデータ20歩分をテストデータとする．
- 被験者間5分割交差検証は，被験者4人分のデータを学習データ，被験者1人分のデータをテストデータとする．

### 問題2
りほちゃんは実験で色々なデータを収集しました．  
実験では，パフをもつ力と皮膚へかけた圧力を収集しました．  
データは3人分です．  
パフは人差し指と親指と中指でもちます．  
パフをあてる場所はおでこ，右頬，左頬です．  
パフの当て方は3種類（長押し，スライド，タッピング）あります．
- りほちゃんはどのような前処理をしたでしょうか？
- りほちゃんはどのような学習をしたでしょうか？
- りほちゃんはどのような検証をしたでしょうか？

- 前処理はせず，指3つの瞬時値から皮膚にかかった圧力の瞬時値をあてる．  
  
  
- 学習器は線形回帰を行った．  
  
  
- まずは場所3つでデータをわけた交差検証

場所3つであけたまま

- 被験者でわけた交差検証
- 被験者内でパフのあてかたでわけた交差検証
- 被験者内でパフのあてかたが同じデータをわけた交差検証
